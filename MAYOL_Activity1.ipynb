```python
def choose_operator(num_choice, num1, num2):
    if num_choice == "1":
        output = num1 + num2
        print(f"{num1} + {num2} = {output}")
    elif num_choice == "2":
        output = num1 - num2
        print(f"{num1} - {num2} = {output}")
    elif num_choice == "3":
        output = num1 * num2
        print(f"{num1} * {num2} = {output}")
    elif num_choice == "4":
        if num2 == 0:
            print(f"\nERROR! Divisor should not be equal to 0.")
        else:
            output = num1 / num2
            print(f"{num1} / {num2} = {output}")
    else:
        print(f"\nINVALID INPUT! Only numbers presented in the choices are permitted.")

def main():
    menu_loop = True
    while menu_loop:
        try:
            num1 = float(input("\nEnter your first number: "))
            num2 = float(input("Enter your second number: "))
            num_choice = input("\nChoose your operation \n[1] Addition [2] Subtraction [3] Multiplication [4] Division \n\nEnter the number of your choice here: ")
            choose_operator(num_choice, num1, num2)
            menu_confirm = input("\nDo you want to try again? \nPress [X] to EXIT, otherwise press [ANY OTHER KEY] to continue...\n")
            if menu_confirm.lower() == "x":
                print("\nThank you!!!")
                menu_loop = False
        except ValueError:
            print("Please enter real numbers only...")
            pass
if __name__ == "__main__":
    main()
```
